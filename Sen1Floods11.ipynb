```python
import pandas as pd
import rasterio
import numpy as np

# Step 1: Load the CSV file
train_df = pd.read_csv('flood_train_data.csv')

# Step 2: Choose a sample file path from the DataFrame (assuming the first column contains paths to S1Hand.tif files)
sample_tif_name = train_df.iloc[0, 0]  # First file path in the first column

# Define the full path in the bucket
bucket_path = f'gs://sen1floods11/v1.1/data/flood_events/HandLabeled/S1Hand/{sample_tif_name}'
local_file_path = f'/content/{sample_tif_name}'

# Step 3: Download the .tif file to your local environment
!gsutil cp {bucket_path} {local_file_path}

# Step 4: Check if the file was downloaded successfully
if os.path.exists(local_file_path):
    print(f"File {local_file_path} downloaded successfully. Proceeding to read data as a DataFrame.")

    try:
        # Open the .tif file with rasterio
        with rasterio.open(local_file_path) as src:
            data = src.read(1)  # Read the first band of the image

            # Convert the data to a DataFrame
            data_df = pd.DataFrame(data)

            # Display a portion of the data for quick inspection (e.g., the first 10 rows)
            print("Data Overview (first 10 rows):")
            print(data_df.head(10))

    except rasterio.errors.RasterioIOError as e:
        print(f"An error occurred while trying to open the .tif file: {e}")
else:
    print(f"File {local_file_path} not found. Please check if the download was successful.")
```

### FUlly understanding TIF format

```python
import rasterio

# Open the TIF file
with rasterio.open(local_file_path) as src:
    print(local_file_path)
    # Retrieve and print metadata
    metadata = src.meta
    print("Metadata:")
    for key, value in metadata.items():
        print(f"{key}: {value}")
```

```python
with rasterio.open(local_file_path) as src:
    # Number of bands in the TIF file
    num_bands = src.count
    print(f"Number of Bands: {num_bands}")

    # Data type of each band
    data_type = src.dtypes[0]  # Usually, all bands have the same data type
    print(f"Data Type: {data_type}")

    # Optionally, check data type of all bands
    print("Data Types for each band:", src.dtypes)
```

```python
with rasterio.open(local_file_path) as src:
    # Print the affine transform for the image (maps pixels to real-world coordinates)
    transform = src.transform
    print("Affine Transform:", transform)

    # Print the coordinate reference system (CRS)
    crs = src.crs
    print("Coordinate Reference System:", crs)
```

```python
# Reading and printing pixel value for each band
with rasterio.open(local_file_path) as src:
    for i in range(1, src.count + 1):  # Band indices start at 1
        # Read the band data as a 2D array
        band_data = src.read(i)

        # Print some sample values
        print(f"Band {i} Sample Data:")
        print(band_data[0:5, 0:5])  # Print a 5x5 sample from the upper-left corner
```

```python
import matplotlib.pyplot as plt
from rasterio.plot import show

with rasterio.open(local_file_path) as src:
    fig, ax = plt.subplots(1, src.count, figsize=(15, 5))

    for i in range(1, src.count + 1):
        data = src.read(i)
        show(data, ax=ax[i-1], cmap='viridis')
        ax[i-1].set_title(f"Band {i}")

    plt.tight_layout()
    plt.show()
```
