## 1 : Gender repartition

```python
#Checking the percentage of applicants based on gender
male_count = len(df[df['gender'] == ' Male'])
female_count = len(df[df['gender'] == ' Female'])
missing_gender = len(df[df['gender'].isnull()])

print('Percentage of Male applicants: ', (male_count/len(df['gender'])*100))
print('percentage of Female applicants: ', (female_count/len(df['gender'])* 100))
print('Percentage of Missing: ', (missing_gender/len(df['gender'])*100))
```

```python
#Creating a visualization 
sns.countplot(x="gender", data=df, palette = 'pastel');
```

## 2: Percentages on some categorical columns

```python
def col_percentages(df_name, col, value1, value2, value3, value4, value5, value6=None):
    """
    This function provides the value counts, percentages and visualizations for columns which 6 values or less.
    """
    print(df_name[col].value_counts())
    print('--------------------------')
    print('\n')
    
    value_1 = len(df_name[df_name[col] == value1])
    value_2 = len(df_name[df_name[col] == value2])
    value_3 = len(df_name[df_name[col] == value3])
    value_4 = len(df_name[df_name[col] == value4])
    value_5 = len(df_name[df_name[col] == value5])
    value_6 = len(df_name[df_name[col] == value6])
    missing_value = len(df_name[df_name[col].isnull()])
    print('Percentage of {} {}: '.format(col, df_name[col].value_counts().index.tolist()[0]), (value_1/len(df_name[col])*100))
    print('percentage of {} {}: '.format(col, df_name[col].value_counts().index.tolist()[1]), (value_2/len(df_name[col])* 100))
    print('Percentage of {} {}: '.format(col, df_name[col].value_counts().index.tolist()[2]), (value_3/len(df_name[col])*100))
    print('Percentage of {} {}: '.format(col, df_name[col].value_counts().index.tolist()[3]), (value_4/len(df_name[col])*100))
    print('Percentage of {} {}: '.format(col, df_name[col].value_counts().index.tolist()[4]), (value_5/len(df_name[col])*100))
    if value6 != None:
        print('percentage of {} {}:'.format(col, df_name[col].value_counts().index.tolist()[5]), (value_6/len(df_name[col])* 100))
    print('Percentage of Missing: ', (missing_value/len(df_name[col])*100))
    
    print('--------------------------')
    print('\n')
    
    sns.countplot(x = col, data=df_name, palette = 'pastel')
    plt.xticks(rotation=45)
```

```python
col_percentages(df,'fam stat', ' Husband', ' Not-in-family', ' Own-child', ' Unmarried', ' Wife', ' Other-relative')
```

```python
col_percentages(df,'race', ' White', ' Black', ' Asian-Pac-Islander', ' Amer-Indian-Eskimo', ' Other')
```
